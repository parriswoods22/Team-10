```python
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from config import key
gmaps.configure(api_key=key)
```

```python
#read in csv of cleaned data
data = pd.read_csv("Team-10/CigaretteData.csv")
data
```

```python
#get unique list of countries

countries = data["Location"].unique()
countries
```

```python
#send list through api to get lat/lng
lats=[]
lngs=[]
count = 0
country_count=len(countries)
for country in countries:
    url = f"https://maps.googleapis.com/maps/api/geocode/json?components=country:{country}&key={key}"
    geo_data = requests.get(url).json()   
    try:
        lat = geo_data["results"][0]["geometry"]["location"]["lat"]
        lng = geo_data["results"][0]["geometry"]["location"]["lng"]
        lats.append(lat)
        lngs.append(lng)
        count = count+1
        print(f"{count}Searching City: {country}")
    except IndexError:
        lat = "Skipped"
        lng = "Skipped"
        lats.append(lat)
        lngs.append(lng)
        count = count+1
        print("City not found")
```
