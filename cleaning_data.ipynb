```python
count=0
rates=[]
for currency in currency_list2:
    # print count to keep track of row
    print(f"Processing {count} of {len(currency_list2)} : {currency}")
    
    # use try statement in case the city does not work in url
    try:
        #get url for each city
        url = f"https://api.exchangerate.host/convert?from={currency}&to=USD"
        api_response = requests.get(url)
        currency_json = api_response.json()
        rates.append(currency_json["info"]["rate"])
    #if error occurs, skip city
    except KeyError:
        print("City not found. Continuing ...")
        rates.append(0)
    
    #add to count to keep track of row
    count=count+1

# final print   
print("-----------------------------------")
print("Data Retrieval Complete")
```
