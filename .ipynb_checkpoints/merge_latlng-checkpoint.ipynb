```python
import pandas as pd

#import csvs
cig = pd.read_csv("CigaretteData.csv")
latlng = pd.read_csv("LatLng.csv")
```

```python
#csvs together
merge=pd.merge(cig, latlng, on="Location")
merge
```

```python
#drop unused index column

final=merge.drop(["Unnamed: 0"],axis=1)
final
```

```python
#export final file to csv

final.to_csv("CigaretteData1", index=True, header=True)
```
